```python
df = pd.read_csv(f"{file_path}/wutheringwaves_character.csv")
```

```python
df.head() #show 5 index
```

```python
# youtube video does the data preparation (fixing the uppercase lowercase for the column)

df.columns = df.columns.str.lower().str.replace(' ','_') # the replace... its like replacing space with underscore

categorical_columns = list(df.dtypes[df.dtypes == 'string'].index) # act as a filter to grab the data names of all the collumns in your dataset that contain text! in this case string!

for c in categorical_columns: # this loop function would format all the data
    df[c]=df[c].str.lower().str.replace(' ','_') # this will like make the data uniform "all on lowercase, and underscore replace the space"
```

```python
df.head()
```

```python
df['character'] = df['character'].str.replace('_-_','_') # format the character collumn data triple underscore to just one underscore
df.head()
```
